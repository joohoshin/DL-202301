보안을 고려하여 별도 파일로 저장을 하고 불러와서 사용합니다. 

````python
import configparser
import openai

config = configparser.ConfigParser()
config.read('/content/drive/MyDrive/Teaching/DL 202301/config.ini')

secret_key = config.get('DEFAULT', 'SECRET_KEY')
organization = config.get('DEFAULT', 'ORGANIZATION')

openai.organization = organization
openai.api_key = secret_key
openai.Model.list()
````

신용카드 설정을 해야 아래 코드가 실행됩니다. 
약간의 무료 사용분이 있으며, 추후 사용량이 넘어가면 비용이 지불됩니다. 

````python
openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "The Los Angeles Dodgers won the World Series in 2020."},
        {"role": "user", "content": "Where was it played?"}
    ]
)
````

한글로 테스트해봅시다. 

참고로 현재 system 메세지는 많이 신경쓰지 않는다고 되어 있습니다. 

이전의 대화를 함께 입력하면 답변이 나오는데 도움이 됩니다. 

````python
res = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=[
        {"role": "user", "content": "너는 언제 데이터까지 학습했어?"},
    ]
)
````

````python
res
````

답변을 출력하면 한글로 제대로 보이게 됩니다. 

````python
res['choices'][0]['message']['content']
````
