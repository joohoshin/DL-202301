모델 파라미터를 위한 자동 미분 기능인 Autograd에 대해 알아봅시다. 

y= 2* x**2 +1 함수에서 x가 2일 때의 미분값을 계산해보겠습니다. 

```python
import torch

# requires_grad = True를 통해서 tensor가 미분값을 계산하도록 설정합니다. 
x = torch.tensor(2.0, requires_grad=True)

y = 2*x**2 + 10  # 수식을 만듭니다. 

y.backward()  # gradient 계산을 합니다. 

print( x.grad)
```

```python
x1 = torch.tensor(2.0, requires_grad=True)
x2 = torch.tensor(3, requires_grad = True)

y = 2*x1**2 + x2 + 10  # 수식을 만듭니다. 

y.backward()  # gradient 계산을 합니다. 

print( x1.grad)
print(x2.grad)
```
