클래스의 변수(Property)과 메서드
* 클래스 변수 : 클래스 내에서 공유되는 값
* 인스턴스 변수수 : 인스턴스마다 가지고 있는 값
* 클래스 메서드 : 클래스 레벨에서 동작하는 메서드
* 인스턴스 메서드 : 인스턴스 레벨에서 동작하는 메서드



```python
class Car:
    # 클래스 변수
    wheels = 4

    # 인스턴스 변수
    def __init__(self, name, color):
        self.name = name
        self.color = color

    # 인스턴스 메서드
    def start(self):
        print(f"{self.color} {self.name} is starting")

    # 클래스 메서드
    @classmethod
    def get_wheels(cls):
        print(f"The car has {cls.wheels} wheels")
```

따라서 클래스 변수의 값을 변경하려면 클래스나 클래스의 인스턴스를 통해 접근하여 변경해야 합니다.

인스턴스에서 변경 시에는 해당 인스턴스의 값만 변경이 됩니다. 

```python
class Car:
    # 클래스 변수
    wheels = 4

    def __init__(self, name, color):
        # 인스턴스 변수
        self.name = name
        self.color = color

# 클래스 변수에 접근하여 변경
print(Car.wheels)  # 4
Car.wheels = 6
print(Car.wheels)  # 6

# 인스턴스를 통해 클래스 변수에 접근하여 변경
car1 = Car("BMW", "Black")
print(car1.wheels) # 6
car1.wheels = 8
print(car1.wheels) # 8
print(Car.wheels)  # 6
```

Private Variable:

* 클래스 내부에서만 활용되는 변수는 2가지로 정의합니다. 
  밑줄 하나를 변수명 앞에 붙입니다. 내부 변수라는 의미로 사용됩니다. 
  실제로는 접근가능하나 관례적으로 내부변수라는 표시입니다. 

* 밑줄 2개 사용시에는 에러가 발생하며, 외부에서 접근할 수 없습니다. 

```python
class Car:
    def __init__(self, name, color):
        self._name = name   # private 변수
        self.__name = name  # private 변수, 외부 접근 안 됨
        self.color = color  # public 변수

car1 = Car("BMW", "Black")
print(car1.color) # Black
print(car1._name)

```

클래스 Subclassing (상속)

만들어진 클래스를 복사하여 사용할 수 있습니다. 
이때 필요한 함수를 새로 정의하거나 추가하여 사용할 수 있습니다

```python
class Vehicle:
    def __init__(self, name, color):
        self.name = name
        self.color = color

    def start(self):
        print(f"{self.color} {self.name} is starting")

class Car(Vehicle):
    def run(self):
        print(f"{self.color} {self.name} is running")

car1 = Car("BMW", "Black")
car1.start() # Black BMW is starting
car1.run()   # Black BMW is running
```

인스턴스 생성시 추가로 작업할게 더 있다면, 
super().__init__()으로 부모 클래스 초기화를 실행한 후에 추가적인 코드 실행되도록 하면 됨

```python
class Vehicle:
    def __init__(self, name, color):
        self.name = name
        self.color = color

class Car(Vehicle):
    def __init__(self, name, color, model):
        super().__init__(name, color)
        self.model = model

car1 = Car("BMW", "Black", "520d")
print(car1.name)   # BMW
print(car1.color)  # Black
print(car1.model)  # 520d
```
